```python
import time
import serial
import io

from pylabnet.utils.logging.logger import LogHandler
```

```python
port = 'COM4'

ser = serial.Serial(
port=port,
baudrate=9600,
timeout=5,
parity=serial.PARITY_NONE,
stopbits=serial.STOPBITS_ONE,
bytesize=serial.EIGHTBITS
)

```

```python
sio = io.TextIOWrapper(io.BufferedRWPair(ser, ser))
```
