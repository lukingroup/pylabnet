```python
import pyvisa
rm = pyvisa.ResourceManager()
rm.list_resources()
```

```python
inst = rm.open_resource("GPIB2::15::INSTR")
print(inst.query("*IDN?"))
inst.write("*RST")
```

```python
inst.write("TSRC 1")   # Trigger Source: External rising
inst.write("TLVL 0.5") # Trigger Level: +0.5 V

# 1st pulse (AB)
inst.write("DLAY 2, 0, 0")      # A has 0 delay after T0
inst.write("DLAY 3, 2, 0.1E-6") # Pulse lasts for 0.1 us
inst.write("LAMP 1, 1")         # Pulse height is 1 V\]  
inst.write("LOFF 1, 0")         # Pulse offset is 0 V
inst.write("LPOL 1, 1")         # Pulse polarity is positive

# 2nd pulse (CD)
inst.write("DLAY 4, 2, 4.68E-6") # C has a delay of 4.68 us after A
inst.write("DLAY 5, 4, 0.1E-6")  # Pulse lasts for 0.1 us
inst.write("LAMP 2, 1")          # Pulse height is 1 V
inst.write("LOFF 2, 0")          # Pulse offset is 0 V
inst.write("LPOL 2, 1")          # Pulse polarity is positive

# Set other pulses to have 0 delay (0 width)
for i in [6, 7, 8, 9]:
    inst.write(f"DLAY {i}, 0, 0")
```

```python
for i in range(6):
    print(inst.query(f"DLAY? {i}"))
```
