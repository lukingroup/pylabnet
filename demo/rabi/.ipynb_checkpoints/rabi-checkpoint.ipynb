```python
mw_src.connect()
```

```python
gated_ctr = gated_ctr.Client(host='localhost', port=3)
```

```python
# gated_ctr.connect()
```

```python
refocus_client = RefocusClient(host='localhost', port=5)
```

## GUI

```python
gui = ComboTraceHMapPBar(trace_title='Average', hm_title='All sweeps')
```

```python
gui.show()
```

## Setup Rabi script instance

```python
my_rabi = Rabi()
```

### Pause server

```python
log_client = LogClient(
    host='localhost', 
    port=0, 
    module_tag='Pause'
)
```

```python
pause_service = PauseService()
pause_service.assign_module(module=my_rabi)
pause_service.assign_logger(logger=log_client)
```

```python
pause_server = GenericServer(
    host='localhost', 
    port=7, 
    service=pause_service
)
```

```python
pause_server.start()
```

```python
pause_server._server.active
```

```python
pause_server._server.clients
```

```python
# my_rabi._is_running
```

### set_hardware()

```python
my_rabi.set_hardware(
    gated_ctr=gated_ctr, 
    mw_src=mw_src, 
    p_gen=p_gen,
    rfcs=refocus_client
)
```

### set_gui()

```python
my_rabi.set_gui(
    trace=gui.trace_fig, 
    heat_map=gui.hm_fig, 
    p_bar=gui.p_bar
)
```

## Measurement

### set_params()

```python
my_rabi.set_params(
    freq=2.893e9,#2.838e9, 
    pwr=-120,
    tau_start=50e-9, 
    tau_stop=2e-6, 
    n_pts=200, 
    n_soft_reps=10, #200 
    n_hrdw_reps=int(5e3), 
    aom_delay=520e-9  #600e-9
)
```

### run()

```python
my_rabi.clear_data()
```

```python
my_rabi.run(rfcs_dur=8)
```
