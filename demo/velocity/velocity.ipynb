```python
def check_operation_status():
    watch_dog = 0
    doubledouble = 0
    while (doubledouble==0):
        single = bool(int(tlb_query(tlb, '*OPC?')))
        time.sleep(1)
        if single:
            doubledouble = bool(int(tlb_query(tlb, '*OPC?')))
            time.sleep(1)
        watch_dog += 1
        if watch_dog > 30:
            break
        
    return doubledouble
```

```python
import numpy as np
```

```python
scan = 3e8/(np.linspace(406.66, 406.7, 15)*1e12)*1e9-0.69

for i, wavelength in enumerate(scan):
    check_passed = check_operation_status()
    if check_passed:
        print('OK to set wavelength')
    else:
        print('Check not passing. Aborting...')
        break
    print(f"Set wavelength to {np.round(wavelength,3)} nm.")
    tlb_query(tlb, f'SOURce:WAVElength {np.round(wavelength,3)}')
    tlb_query(tlb, 'OUTPut:TRACK 1')
```
